```python
import pandas as pd  # visualizacao e manipulacao do dataframe
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
alldata = pd.read_csv('/home/thiago/Downloads/creditcard.csv')

print('O arquivo inteiro tem {} linhas e {} colunas'.format(alldata.shape[0],alldata.shape[1]))

alldata.iloc[0:100] 
```

```python
X = alldata.drop('Class', axis=1)  
y = alldata['Class']  

from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20) #define quantidade de dados que sera usada para teste
print('data splitted sucessfully')

#ferramentas para suport vector machines
from sklearn.svm import SVC 
#avaliacao dos resultados da previsao
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
svclassifier = SVC(kernel='linear') 
print('iniciando treinamento do modelo linear')
svclassifier.fit(X_train, y_train) 
print('linear model trained sucessfully')

#com o modelo de treino sao feitas previsoes para o modelo de teste
y_pred = svclassifier.predict(X_test) 
print('previsao baseada no modelo linear realizada' )


linear_SVM_CFmatrix = confusion_matrix(y_test,y_pred)
print(linear_SVM_CFmatrix)

linear_SVM_CReport = classification_report(y_test,y_pred)

print(linear_SVM_CReport)  
```

```python
#svm polinomial
poly_svm_CFmatrix = []
poly_svm_Creport = []
for i in range(2, 5)
    print('iniciando modelo SVM polinomial')
    print('')
    print('grau polinomial do modelo sera ', i)
    print('')

    #grau do polinomio usado pelo svm
    svclassifier = SVC(kernel = 'poly', degree = i)  
    svclassifier.fit(X_train, y_train) 
    print('polynomial SVM model trained sucessfully')
    print('')
    
    #fazer previsao
    y_pred = svclassifier.predict(X_test) 
    print('polynomial svm based prediction made sucessfully for poly = ', i )
    
    
    #resultados da previsao
    print(confusion_matrix(y_test, y_pred))
    poly_svm_predictions.append(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))  
    poly_svm_Creport.append(classification_report(y_test, y_pred))

print(poly_svm_CFmatrix)
print(poly_svm_Creport)
```

```python
#modelo gaussiano
print('iniciando modelo de SVM gaussiano')
svclassifier = SVC(kernel='rbf')

print('')

print('iniciando treinamento do modelo de SVM gaussiano')

#treinamento do modelo SVM gaussiano
svclassifier.fit(X_train, y_train)
print('modelo treinado com sucesso')

y_pred = svclassifier.predict(X_test) 
print('previsao realizada com sucesso')

gaussian_SVM_CFmatrix = confusion_matrix(y_test, y_pred)
gaussian_SVM_Creport = classification_report(y_test, y_pred)
print(gaussian_SVM_CFmatrix)  
print(gaussian_SVM_Creport)
```

```python
#modelo SVM sigmoide
print('iniciando modelo de SVM sigmoide')
svclassifier = SVC(kernel='sigmoid')

print('')

print('iniciando treinamento do modelo de SVM sigmoide')
#treinamento do modelo SVM sigmoide
svclassifier.fit(X_train, y_train)
print('modelo treinado com sucesso')

y_pred = svclassifier.predict(X_test) 
print('previsao realizada com sucesso')



sigmoid_SVM_CFmatrix = confusion_matrix(y_test, y_pred)
sigmoid_SVM_Creport = classification_report(y_test, y_pred)
print(sigmoid_SVM_CFmatrix)  
print(sigmoid_SVM_Creport)
```
